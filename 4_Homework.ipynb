## 2 задача
### 2.1 Изучите распределение домов от наличия вида на набережную
Постройте график

Сделайте выводы

```python
plt.pie(df['waterfront'].value_counts().values, autopct= '%1.1f%%', labels= df['waterfront'].value_counts().values);
```

### 2.2 Изучите распределение этажей домов


```python
floors_houses = df['floors'].value_counts()
index_values = floors_houses.values
index_name = floors_houses.index
plt.bar(index_name, index_values)
plt.title('распределение этажей домов')
plt.xlabel('Этажность')
plt.ylabel('Колличество');
```

### 2.2 Изучите распределение состояния домов

```python
condition_houses = df['condition'].value_counts()
index_values = condition_houses.values
index_name = condition_houses.index
plt.bar(index_name, index_values)
plt.title('распределение состояния домов')
plt.ylabel('колличество');
```
